VAE - Variational AutoEncodes (вариационный автокодировщик)

```python
import keras
from keras import layers
from keras import backend as K
from keras.models import Model
from keras.datasets import mnist

import numpy as np

import matplotlib.pyplot as plt

from scipy.stats import norm
```

https://github.com/fchollet/deep-learning-with-python-notebooks/blob/master/8.4-generating-images-with-vaes.ipynb

Простая нейросеть отображающая входное изображение x в два вектора z_mean, z_log_var. 

```python
img_shape = (28, 28, 1)
batch_size = 16
latent_dim = 2 #Размерность скрытого простарнства: двумерная плоскость

input_img = keras.Input(shape = img_shape)
x = layers.Conv2D(32, 3, padding = 'same', activation = 'relu')(input_img)
x = layers.Conv2D(64, 3, padding = 'same', activation = 'relu', strides = (2, 2))(x)
x = layers.Conv2D(64, 3, padding = 'same', activation = 'relu')(x)
x = layers.Conv2D(64, 3, padding = 'same', activation = 'relu')(x)

shape_before_flattening = K.int_shape(x)

x = layers.Flatten()(x)
x = layers.Dense(32, activation = 'relu')(x)

z_mean = layers.Dense(latent_dim)(x)
z_log_var = layers.Dense(latent_dim)(x)
```

Скрытое пространсво итоговых точек подчиняется распределению Гаусса. Значения входного изображения превращаются в параметры статистического распределения: среднее и дисперсию.

```python
def sampling(args):
    '''
    Функция выбора точки из скрытого пространства
    '''
    z_mean, z_log_var = args
    epsilon = K.random_normal(shape = (K.shape(z_mean)[0], latent_dim), mean = 0., stddev = 1.)  #Тензор смещения
    return z_mean + K.exp(z_log_var) * epsilon
```

```python
z = layers.Lambda(sampling)([z_mean, z_log_var]) 
```

Вторая часть нейросети - декодер. Отображает точки из срытого простарнства в изображения

```python
decoder_input = layers.Input(K.int_shape(z)[1:])

x = layers.Dense(np.prod(shape_before_flattening[1:]), activation = 'relu')(decoder_input) #Увеличиваем разрешение входа до размера исходного изображения в энкодере после нескольких Conv2D и Flatten()
x = layers.Reshape(shape_before_flattering[1:])(x)

x = layers.Conv2DTranspose(32, 3, padding = 'same', activation = 'relu', strides = (2, 2))(x)
x = layers.Conv2D(1, 3, padding = 'same', activation = 'sigmoid')(x)

decoder = Model(decoder_input, x) #Создаем модель, преобразующаю decoder_input -> x (декодированное изображение)
```

```python
z_decoded = decoder(z)
```

Собственный слой для вычисления потерь VAE

```python
class CustomVariationallLayer(keras.layers.Layer):
    
    def vae_loss(self, x, z_decoded):
        x = K.flatten(x)
        z_decoded = K.flatten(z_decoded)
        xent_loss = keras.metrics.binary_crossentropy(x, z_decoded)
        kl_loss = -5e-4 * K.mean(1 + z_log_var - K.square(z_mean) - K.exp(z_log_var), axis = -1)
        return K.mean(xent_loss + kl_loss)
    
    def call(self, inputs):
        x = inputs[0]
        z_decoded = inputs[1]
        loss = self.vae_loss(x, z_decoded)
        self.add_loss(loss, inputs = inputs)
        return x
```

```python
y = CustomVariationallLayer()([input_img, z_decoded])
```

Теперь можно обучить модель. Поскольку потери зашиты в отдельный слой, то функцию потерь при компиляции не указываем (loss = None).
Так же случае fit() y = None

```python
vae = Model(input_img, y)
vae.compile(optimizer = 'rmsprop', loss = None)
vae.summary()

(x_train, _), (x_test, y_test) = mnist.load_data()

x_train = x_train.astype('float32') / 255
x_train = x_train.reshape(x_train.shape + (1,))

x_test = x_test.astype('float32') / 255
x_test = x_test.reshape(x_test.shape + (1,))

vae.fit(x = x_train, y = None, 
        shuffle = True, epochs = 10,
        batch_size = batch_size,
        validation_data = (x_test, None))
```

Отобразим результаты

```python
n = 15  # размер итоговой сетки изображения цифр
digit_size = 28
figure = np.zeros((digit_size * n, digit_size * n))

# https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.norm.html
# Сетка значений линейного пространства в гауссовом
grid_x = norm.ppf(np.linspace(0.05, 0.95, n))
grid_y = norm.ppf(np.linspace(0.05, 0.95, n))

for i, yi in enumerate(grid_x):
    for j, xi in enumerate(grid_y):
        z_sample = np.array([[xi, yi]])
        z_sample = np.tile(z_sample, batch_size).reshape(batch_size, 2)
        x_decoded = decoder.predict(z_sample, batch_size=batch_size)
        digit = x_decoded[0].reshape(digit_size, digit_size)
        figure[i * digit_size: (i + 1) * digit_size,
               j * digit_size: (j + 1) * digit_size] = digit

plt.figure(figsize=(10, 10))
plt.imshow(figure, cmap='Greys_r')
plt.show()
```

Мы получили непрерывное распределение разных классов цифр, где одна цифра превращается в другую по пути через скрытое пространство.
